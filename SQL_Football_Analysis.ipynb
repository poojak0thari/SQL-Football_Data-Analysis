DATA INSIGHTS HUNT

1. Let's see the average rating, averge potential and total penalties of each player in the matches played between year 2016-2011.

```python
I1= pd.read_sql(""" select 	Player.player_api_id, Player.player_name , avg(PA.overall_rating) as avg_rating,
                    avg(PA.potential) as avg_potential, sum(PA.penalties) as total_penalties
                    from Player
                    left join Player_Attributes as PA
                    on Player.player_api_id= PA.player_api_id
                    where PA.date between '2011-01-01' and '2016-12-31'
                    group by PA.player_api_id
                    order by avg_rating desc;""", conn)
I1
```

This helps us understand the top performers in the last five year from 2011-2016.

2. NOW LET'S FIND OUT THE PERFORMANCE OF EACH TEAM WHO PLAYED FOR SEASON 2015/2016.

```python
a= pd.read_sql(" PRAGMA table_info(Match); ", conn)
a.head(20)
```

```python
I2=pd.read_sql(""" select match_api_id, team_long_name , count(W.winner) as number_of_wins
                  FROM (select  match_api_id, case
                  when (home_team_goal- away_team_goal) >0 then home_team_api_id
                  when (home_team_goal- away_team_goal)=0 then 0
                  else away_team_api_id
                  end as winner,  T.team_long_name
                  from Match
                  join Team as T
                  on winner= T.team_api_id
                 where season= '2015/2016') AS W
                 group by team_long_name
                 order by number_of_wins desc""", conn)
I2
```

So we can conclude that in the Season 2015/2016 the team Paris Saint-Germain won 30 matches , hence it is the best performed team of this season.
